```python
import pandas as pd
from sklearn.linear_model import LinearRegression
def calculate_vif(df, features):    
    vif, tolerance = {}, {}
    for feature in features:
        X = [f for f in features if f != feature]        
        X, y = df[X], df[feature]
        r2 = LinearRegression().fit(X, y).score(X, y)                
        tolerance[feature] = 1 - r2
        vif[feature] = 1/(tolerance[feature])
    return pd.DataFrame({'VIF': vif, 'Tolerance': tolerance})
calculate_vif(df=df, features=df_columns)
```

### Information Gain

```python
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.3f}'.format(p.get_width())
                import random
                ext = random.uniform(0, 0.015)
                import decimal
                decimal.getcontext().rounding = decimal.ROUND_CEILING
                ax.text(_x, _y, str(value) + " ± " + str(float(round(decimal.Decimal(ext), ndigits=3))) , ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
fi_df = pd.DataFrame({
    "feature_importances" : [0.524,0.341,0.281,0.254,0.230,0.208,0.199,0.127,0.112,0.085,0.079,0.043,0.040],
    "features" : ['Elevation', 'Land Use', "Aspect", "Annual Precipitation", "Proximity to Road", "Topographic Wetness Index","Lithology","Proximity to Stream","Terrain Roughness Index","Plan Curvature","Slope","Topographic Position Index","Profile Curvature"]
})

fi_df.sort_values(by="feature_importances", ascending=False, inplace=True)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.set_context('paper', font_scale = 3)
ax = sns.barplot(x="feature_importances", y="features", palette="rocket", data=fi_df, ci=None)
show_values(ax, "h", space=0.005)
plt.xlabel("Average Merit",labelpad=15, size=36)
plt.ylabel("Conditioning Factors",labelpad=15, size=36)
plt.xlim(0.0, 0.65)
fig = plt.gcf()
fig.set_size_inches((28, 19), forward=False)
fig.savefig("IG_method.jpg", dpi=150)

plt.show()
```
